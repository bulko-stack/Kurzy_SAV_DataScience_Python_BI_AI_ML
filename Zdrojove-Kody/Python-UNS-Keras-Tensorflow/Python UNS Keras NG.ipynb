# ⚒️ Top knižnice pre UNS a Machine Learning 

1. **🧠 Keras**: Knižnica pre Python, ktorá sa používa na vytváranie a trénovanie neurónových sietí. Je zameraná na jednoduché a rýchle prototypovanie a je často používaná ako vysokú **úroveň nad TensorFlow**.

2. **⚙️ TensorFlow**: Otvorená knižnica od **Google** pre numerické výpočty a strojové učenie. Je obzvlášť populárna pre hlboké učenie a podporuje rozsiahle nasadenie na rôznych platformách.
   
3. **🤖 scikit-learn**: Knižnica pre Python, ktorá poskytuje **jednoduché** a **efektívne** **nástroje** na analýzu dát a strojové učenie. Obsahuje rôzne algoritmy pre klasifikáciu, regresiu a zhlukovanie.
   
4. **🐍 PyTorch**: Otvorená knižnica od **Facebooku** pre hlboké učenie. Umožňuje dynamické výpočtové grafy a je známa svojou intuitívnosťou a jednoduchosťou pri vývoji komplexných modelov.
   
5. **🎨 Fastai**: Knižnica **založená** na **PyTorch**, ktorá zjednodušuje proces vytvárania a školenia modelov hlbokého učenia. Je orientovaná na praktické využitie a rýchle prototypovanie.
   
6. **🔍 XGBoost**: Efektívna a rýchla knižnica na **gradient** **boosting**. Je široko používaná v súťažiach a aplikáciách strojového učenia, hlavne na úlohy klasifikácie a regresie.
   
7. **📈 LightGBM**: Knižnica od **Microsoftu** pre gradient boosting, ktorá je optimalizovaná na výkon a rýchlosť. Umožňuje spracovanie veľkých dát a je obľúbená pre svoje rýchle trénovanie.
   
8.  **🔗 CatBoost**: Knižnica od **Yandexu**, ktorá je špeciálne navrhnutá na prácu s kategorizovanými údajmi. Je známa svojou jednoduchosťou a vysokou presnosťou.
   
9. **🌊 H2O.ai**: Platforma pre automatizované strojové učenie, ktorá umožňuje rýchle vytváranie a nasadenie modelov. Podporuje viacero jazykov a má robustné rozhranie pre vizualizáciu.
    
10. **🛠️ MLflow**: Nástroj pre spravovanie životného cyklu strojového učenia. Umožňuje sledovanie experimentov, spravovanie modelov a ich nasadenie do produkcie.

## 🧠 Čo je Keras?
- Keras je vysokoúrovňová knižnica pre tvorbu a trénovanie neurónových sietí v jazyku Python
- Umožňuje rýchly prototyp modelov hlbokého učenia vďaka jednoduchej a čitateľnej syntaxe
- Beží nad výpočtovým backendom ako je TensorFlow, ktorý zabezpečuje efektívne výpočty (CPU/GPU)
- Od verzie TensorFlow 2.x je Keras súčasťou TensorFlow ako tf.keras
  
**👨‍💻 Autor**: François Chollet, výskumník umelej inteligencie v Google 2015  
**🏢 Miesto**: Google (projekt Theano → TensorFlow)  
**📜 Licencia**: Open Source (MIT)  

### Hlavné Vlastnosti Keras
1. ✅ **Jednoduchosť** – intuitívne API vhodné pre začiatočníkov aj výskumníkov
1. 🔁 **Modularita** – modely sú skladané z vrstiev ako stavebnice
2. 🔍 **Podpora pre výskum aj prax** – vhodné pre rýchle experimenty aj nasadenie do produkcie
3. 🌍 **Podpora viacerých platforiem** – kompatibilita s TensorFlow, GPU, TPU, mobilmi (TF Lite)

## ⚙️ Čo je Tensorflow?
- TensorFlow je open-source knižnica na numerické výpočty a strojové učenie
- Využíva graf výpočtov (dataflow graph) pre efektívne spracovanie údajov
- Podporuje neurónové siete, regresie, klasifikácie, ale aj všeobecné matematické operácie
- Funguje na CPU, GPU aj TPU, ideálny pre škálovateľné ML aplikácie

**👨‍💻 Autor:** tím Google Brain (hlavne Jeff Dean a Rajat Monga)  
**🏢 Miesto:** Google Brain, USA  
**📜 Licencia:** Open Source (Apache License 2.0)  

###  Hlavné Vlastnosti TensorFlow
1. **🧮 Podpora pre výpočty s tenzormi** – základná dátová jednotka je Tensor
2. **🛠️ Flexibilita** – tvorba jednoduchých aj komplexných modelov  
3. **🚀 Vysoký výkon** – optimalizácia pre paralelizmus (CPU, GPU, TPU)  
4. **🧰 Bohatá ekosystémová knižnica** – napr. TensorBoard, TF Lite, TF Serving, TF Hub  
5. **🌐 Nasadenie na rôzne platformy** – desktop, mobil, cloud, embedded zariadenia  
6. **📈 Vizualizácia a ladenie** – pomocou TensorBoard  
7. **🔗 Integrácia s Keras** – Keras je oficiálne high-level API TensorFlow

### 🧠 Význam kriviek v grafe:  
**🔵 Modrá krivka: Strata (loss)**

- Váš graf ukazuje, že strata sa znižuje, čo je dobrý znak.

**🟢 Zelená krivka: Presnosť (accuracy)**  
- Ukazuje percento správne klasifikovaných príkladov.
- Stúpajúca hodnota znamená, že model sa zlepšuje.
- Váš model sa zlepšoval v presnosti počas všetkých 10 epoch.

✅ Čo to znamená pre vás:
Model sa učí dobre, strata klesá a presnosť rastie.

#  Vytvorenie Modelu

# 🎓 Vytvorenie a tréning jednoduchého modelu
- Základná štruktúra neurónovej siete v Keras (Sequential model)
- Pridávanie vrstiev (Dense, Activation) a nastavenie optimalizátorov
- Tréning modelu na jednoduchých dátach (napr. MNIST)
- Použitie metód fit() a evaluate()

## 1️⃣ Import knižníc
**📌 Úloha:** Importovať základné knižnice potrebné pre vytváranie a trénovanie neurónových sietí v Keras.

- TensorFlow je framework na prácu s neurónovými sieťami.
- Keras je jeho high-level API vhodné pre začiatočníkov.
- Dataset MNIST obsahuje 70 000 čiernobielych obrázkov číslic 0–9.

## 2️⃣ Načítanie a predspracovanie dát (MNIST)
**📌 Úloha:** Načítať dataset MNIST, normalizovať pixely a previesť výstupy do one-hot kódovania.

- Hodnoty pixelov sú v rozsahu 0 až 255 – normalizujeme ich na 0 až 1.
- Výstupné triedy (0–9) musíme konvertovať na vektory dĺžky 10.

## 3️⃣ Vytvorenie modelu – základná štruktúra
**📌 Úloha:** Vytvoriť jednoduchý sekvenčný model pomocou Sequential, ktorý obsahuje skrytú a výstupnú vrstvu.

- Flatten prevádza 2D obraz (28x28) na 1D vektor (784).
- Dense vrstva má plné spojenie s predchádzajúcou vrstvou.
- ReLU aktivácia sa používa pre nelinearitu.
- Softmax zabezpečí výstup v podobe pravdepodobností pre každú triedu.

## 4️⃣ Kompilácia modelu
**📌 Úloha:** Nastaviť optimalizátor, funkciu straty a metriky hodnotenia výkonnosti modelu.

- Adam je adaptívny optimalizátor vhodný pre väčšinu úloh.
- categorical_crossentropy je strata pre viac-triedovú klasifikáciu.
- accuracy sleduje podiel správnych predikcií.

## 5️⃣ Tréning modelu – metóda fit()
**📌 Úloha:** Vytrénovať model na trénovacích dátach pomocou metódy fit() a uložiť výsledok do premennej history.

- epochs – koľkokrát prejdeme celé trénovacie dáta.
- batch_size – počet vzoriek, ktoré sa spracúvajú naraz.
- validation_split – časť trénovacích dát sa použije ako validačná množina.

## 6️⃣ Vyhodnotenie modelu – metóda evaluate()
**📌 Úloha:** Vyhodnotiť model na testovacích dátach a zobraziť presnosť klasifikácie.

- evaluate() vracia hodnotu straty a presnosti pre nové (nevidené) dáta. 

## 7️⃣ Vizualizácia priebehu trénovania pomocou Matplotlib
**📌 Úloha:** Zobraziť graf trénovacej chyby (loss) a presnosti (accuracy) po jednotlivých epochách trénovania modelu v Keras.

- V grafe porovnáme trénovacie a validačné výsledky.
- Pomáha vizualizovať, či model napríklad nepreučený (overfitting).

## 8️⃣ Bonusová úloha na precvičenie
**📌 Úloha:** Otestovať rôzne konfigurácie modelu a sledovať vplyv na výslednú presnosť.

🎯 Skúste nasledovné zmeny a opäť spustite tréning:

- Zmeniť počet epôch: epochs=10
