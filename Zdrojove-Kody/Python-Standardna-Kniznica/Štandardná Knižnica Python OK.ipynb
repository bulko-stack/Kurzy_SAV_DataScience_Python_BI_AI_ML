```python
import os
import sys
import importlib.util

# Funkcia na naƒç√≠tanie ≈°tandardn√Ωch kni≈æn√≠c
def get_standard_modules():
    # Z√≠skanie cesty k ≈°tandardnej kni≈ænici Pythonu
    standard_lib = os.path.join(sys.prefix, 'lib', f'python{sys.version_info[0]}.{sys.version_info[1]}', 'site-packages')

    modules = []
    # Prehƒæad√°vanie kni≈æn√≠c v ≈°tandardnej kni≈ænici
    for root, dirs, files in os.walk(standard_lib):
        for file in files:
            if file.endswith(".py"):  # Kontrola s√∫borov s pr√≠ponou .py
                module_name = file[:-3]  # Odstr√°nenie pr√≠pony .py
                modules.append(module_name)
    return modules

# Z√≠skanie zoznamu ≈°tandardn√Ωch modulov
modules = get_standard_modules()

print(f"Pocet modulov: {len(modules)}") 

print(f"Zoznam modulov: {modules}") 
```

### Funkcia all() 
- Vracia True, ak v≈°etky polo≈æky v iterovateƒænom objekte s√∫ pravdiv√© (alebo ak je iterovateƒæn√Ω objekt pr√°zdny).
- Pou≈æ√≠va na overenie, ƒçi s√∫ v≈°etky prvky v iterovateƒænom objekte (napr. zoznam, mno≈æina, atƒè.) pravdiv√©. Ak s√∫ v≈°etky hodnoty pravdiv√© (tzn. hodnoty, ktor√© sa v kontexte Pythonu hodnotia ako True), funkcia vr√°ti hodnotu True. 
- Ak aspo≈à jedna hodnota v iterovateƒænom objekte je nepravdiv√° (tzn. hodnoty ako False, None, pr√°zdne re≈•azce, ƒç√≠sla 0, atƒè.), funkcia vr√°ti False.
- Veƒæmi u≈æitoƒçn√° v pr√≠padoch, kde chcete skontrolova≈•, ƒçi v≈°etky prvky spƒ∫≈àaj√∫ urƒçit√∫ podmienku, bez toho, aby ste museli iterova≈• cez ka≈æd√Ω prvok manu√°lne.

**V√Ωhody:**
1. Umo≈æ≈àuje jednoducho a efekt√≠vne skontrolova≈• pravdivos≈• viacer√Ωch podmienok naraz.
2. Zjednodu≈°uje z√°pis k√≥du, preto≈æe sa nemus√≠ pou≈æ√≠va≈• cyklus alebo zlo≈æit√© podmienky.

```python
# Zisti≈•, ƒçi v≈°etky prvky v zozname s√∫ pravdiv√©:
# Pr√≠klad pre funkciu all()
zoznam = [True, True, False]
vysledok = all(zoznam)
print(vysledok)  # V√Ωstup: False

# Pr√°zdny zoznam
empty_list = []
print(all(empty_list))  # V√Ωstup: True
```

```python
# Skontrolova≈•, ƒçi v≈°etky hodnoty v zozname s√∫ kladn√© ƒç√≠sla:

numbers = [1, 2, 3, 4]
print(all(n > 0 for n in numbers))  # Vr√°ti True, preto≈æe v≈°etky ƒç√≠sla s√∫ kladn√©
```

```python
# Skontrolova≈•, ƒçi v≈°etky re≈•azce v zozname nie s√∫ pr√°zdne:

strings = ['apple', 'banana', 'cherry']
print(all(s != '' for s in strings))  # Vr√°ti True, preto≈æe ≈æiadny re≈•azec nie je pr√°zdny
```

### Funkcia any() 
- Vracia True, ak aspo≈à jedna polo≈æka v iterovateƒænom objekte je pravdiv√°.
- Pou≈æ√≠va na overenie, ƒçi aspo≈à jeden prvok v iterovateƒænom objekte (napr. zoznam, mno≈æina, atƒè.) je pravdiv√Ω. Ak aspo≈à jeden prvok hodnot√≠ ako True (tzn. nie je False, None, pr√°zdny re≈•azec, ƒç√≠slo 0, atƒè.), funkcia vr√°ti hodnotu True. 
- Ak v≈°etky prvky s√∫ nepravdiv√© (tzn. hodnoty ako False, None, pr√°zdne re≈•azce, ƒç√≠sla 0, atƒè.), funkcia vr√°ti False.
- Veƒæmi u≈æitoƒçn√°, keƒè potrebujete vedie≈•, ƒçi niektor√Ω prvok spƒ∫≈àa urƒçit√∫ podmienku, bez toho, aby ste museli prehƒæad√°va≈• cel√Ω objekt.

**V√Ωhody:**
1. Funkcia any() je ide√°lna na r√Ωchlu kontrolu, ƒçi existuje aspo≈à jeden pravdiv√Ω prvok v iterovateƒænom objekte.
2. Umo≈æ≈àuje urobi≈• t√∫to kontrolu bez potreby cyklov alebo zlo≈æit√Ωch podmienok.

```python
# Zisti≈•, ƒçi aspo≈à jedna hodnota v zozname je True:

# Pr√≠klad pre funkciu any()
zoznam = [False, False, True]
vysledok = any(zoznam)
print(vysledok)  # V√Ωstup: True

# V≈°etky polo≈æky s√∫ False
empty_list = [False, False, False]
print(any(empty_list))  # V√Ωstup: False
```

```python
# Skontrolova≈•, ƒçi aspo≈à jedno ƒç√≠slo v zozname je kladn√©:

numbers = [-1, -2, 3, -4]
print(any(n > 0 for n in numbers))  # Vr√°ti True, preto≈æe ƒç√≠slo 3 je kladn√©
```

```python
# Skontrolova≈•, ƒçi aspo≈à jeden re≈•azec v zozname nie je pr√°zdny:

strings = ['', '', 'cherry']
print(any(s != '' for s in strings))  # Vr√°ti True, preto≈æe 'cherry' nie je pr√°zdny
```

### Funkcia bytearray() 
- Vytv√°ra objekt typu bytearray, ktor√Ω je mutable (modifikovateƒæn√Ω) a sl√∫≈æi na ukladanie bajtov.
- Pou≈æ√≠va na vytvorenie mutable (meniteƒæn√©ho) typu d√°t, ktor√Ω reprezentuje sekvenciu bajtov. 
- Tento typ d√°t je podobn√Ω obyƒçajn√©mu typu bytes, ale na rozdiel od neho umo≈æ≈àuje upravova≈• (modifikova≈•) obsah, ƒço znamen√°, ≈æe je mo≈æn√© prid√°va≈•, meni≈• alebo maza≈• bajty.
- Ide√°lna na manipul√°ciu s bin√°rnymi d√°tami, kde potrebujete flexibilitu a v√Ωkon pri modifik√°cii bajtov v pam√§ti.

**V√Ωhody:**
1. Mutable: Na rozdiel od bytes, ktor√© s√∫ nemenn√©, bytearray umo≈æ≈àuje priamu manipul√°ciu s obsahom.
2. Efekt√≠vnos≈•: Je to efekt√≠vny sp√¥sob na pr√°cu s bin√°rnymi d√°tami alebo k√≥dovan√Ωmi re≈•azcami, ktor√© sa potrebuj√∫ meni≈•.
3. Kompatibilita: Pou≈æ√≠va sa na pr√°cu s bin√°rnymi s√∫bormi alebo komunik√°ciou cez protokoly, kde je potrebn√© prisp√¥sobova≈• bajty.

```python
# Vytvorenie pr√°zdneho bytearray:

b = bytearray()
print(b)  # Vytlaƒç√≠: bytearray(b'')
```

```python
# Vytvorenie bytearray zo zoznamu ƒç√≠sel:

# Pr√≠klad pre funkciu bytearray()
bajty = bytearray([65, 66, 67])  # ASCII hodnoty pre 'A', 'B', 'C'
print(bajty)  # V√Ωstup: bytearray(b'ABC')

# Modifik√°cia
bajty[0] = 90  # Zmena prv√©ho znaku na 'Z'
print(bajty)  # V√Ωstup: bytearray(b'ZBC')
```

```python
# Vytvorenie bytearray z re≈•azca pomocou k√≥dovania:
b = bytearray('hello', 'utf-8')
print(b)  # Vytlaƒç√≠: bytearray(b'hello')
```

### Funkcia callable() 
- Skontroluje, ƒçi je objekt volateƒæn√Ω (t. j. ƒçi m√¥≈æe by≈• zavolan√Ω ako funkcia).
- Pou≈æ√≠va na zistenie, ƒçi je objekt volateƒæn√Ω, tzn. ƒçi sa d√° zavola≈• ako funkcia alebo met√≥da. 
- Funkcia callable() vr√°ti hodnotu True, ak je objekt volateƒæn√Ω (napr√≠klad funkcia, met√≥da, trieda s met√≥dou __call__), a False, ak nie je.
- Umo≈æ≈àuje flexibilne zisti≈•, ƒçi objekt podporuje volanie, ƒço je u≈æitoƒçn√© pri pr√°ci s funkciami, met√≥dami alebo in√Ωmi objekty, ktor√© implementuj√∫ met√≥du __call__.

**V√Ωhody:**
1. Funkcia callable() je u≈æitoƒçn√° na dynamick√∫ kontrolu, ƒçi je objekt funkciou alebo met√≥dou, a teda ho m√¥≈æeme zavola≈•.
2. Je to praktick√© pri pr√°ci s objektmi, ktor√© m√¥≈æu by≈• r√¥zneho typu a nie v≈ædy je zaruƒçen√©, ≈æe s√∫ volateƒæn√©.

```python
# Skontrolova≈•, ƒçi je funkcia volateƒæn√°:

# Pr√≠klad pre funkciu callable()
def funkcia():
    pass

objekt = 123

print(callable(funkcia))  # V√Ωstup: True
print(callable(objekt))   # V√Ωstup: False
```

```python
# Skontrolova≈•, ƒçi je obyƒçajn√Ω objekt volateƒæn√Ω (nie je)
my_string = "Hello, World!"
print(callable(my_string))  # Vr√°ti False, preto≈æe re≈•azec nie je volateƒæn√Ω
```

```python
# Skontrolova≈•, ƒçi je trieda volateƒæn√° (ak m√° met√≥du __call__):

class MyClass:
    def __call__(self):
        return "Som volatelna Trieda (callable)!"

obj = MyClass()
print(callable(obj))  # Vr√°ti True, preto≈æe MyClass m√° met√≥du __call__
```

### Funkcia complex() 
- Vytv√°ra komplexn√© ƒç√≠slo. M√¥≈æe prija≈• re√°lnu a imagin√°rnu ƒças≈• ako argumenty.
- Pou≈æ√≠va sa na vytvorenie komplexn√©ho ƒç√≠sla. 
- Komplexn√© ƒç√≠slo je ƒç√≠slo, ktor√© obsahuje re√°lnu a imagin√°rnu ƒças≈•, a je vo forme a + bj, kde a je re√°lna ƒças≈• a b je imagin√°rna ƒças≈•, priƒçom j je imagin√°rna jednotka (v Pythone je to j, nie i, ako je be≈æn√© v matematike).
- U≈æitoƒçn√° pri pr√°ci s komplexn√Ωmi ƒç√≠slami v Pythone a je nevyhnutn√° na ich tvorbu a manipul√°ciu.

**V√Ωhody:**
1. Funkcia complex() umo≈æ≈àuje jednoducho vytvori≈• komplexn√© ƒç√≠sla, ktor√© s√∫ d√¥le≈æit√© v r√¥znych oblastiach matematiky, in≈æinierstva a veda (napr. v anal√Ωze sign√°lov, Fourierovej anal√Ωze, kvantovej fyzike, atƒè.).
2. Python automaticky podporuje aritmetiku s komplexn√Ωmi ƒç√≠slami, vr√°tane sƒç√≠ta≈•, odƒç√≠ta≈•, n√°sobi≈• a deli≈• komplexn√© ƒç√≠sla.

```python
# Vytvorenie komplexn√©ho ƒç√≠sla z re√°lnej a imagin√°rnej ƒçasti:

c = complex(2, 3)
print(c)  # Vytlaƒç√≠: (2+3j)
```

```python
# Vytvorenie komplexn√©ho ƒç√≠sla len z re√°lnej ƒçasti (imagin√°rna ƒças≈• je 0):

c = complex(4)
print(c)  # Vytlaƒç√≠: (4+0j)
```

```python
# Vytvorenie komplexn√©ho ƒç√≠sla z re≈•azca (dohodnut√Ω form√°t):

c = complex('2+3j')
print(c)  # Vytlaƒç√≠: (2+3j)
```

```python
# Z√≠skanie re√°lnej a imagin√°rnej ƒçasti komplexn√©ho ƒç√≠sla:

c = complex(3, 4)
print(c.real)  # Vytlaƒç√≠: 3.0
print(c.imag)  # Vytlaƒç√≠: 4.0
```

### Funkcia delattr() 
- Sl√∫≈æi na vymazanie atrib√∫tu objektu.
- Pou≈æ√≠va na odstr√°nenie (vymazanie) atrib√∫tu z objektu. 
- Funkcia vy≈æaduje 2 argumenty: objekt a re≈•azec obsahuj√∫ci n√°zov atrib√∫tu, ktor√Ω sa m√° odstr√°ni≈•.
- Funkcia delattr() je v√Ωkonn√Ω n√°stroj na dynamick√© odstr√°nenie atrib√∫tov z objektov v Pythonu.

**V√Ωhody:**
1. Funkcia delattr() je u≈æitoƒçn√°, keƒè potrebujete dynamicky upravi≈• objekt a odstr√°ni≈• atrib√∫ty poƒças behu programu.
2. Je to efekt√≠vny sp√¥sob, ako vymaza≈• atrib√∫ty, ak u≈æ nie s√∫ potrebn√©, ƒç√≠m ≈°etr√≠te pam√§≈• a udr≈æiavate objekt v ƒçistej≈°om stave.




```python
# Odstr√°nenie atrib√∫tu z objektu:

class Osoba:
    def __init__(self, meno, vek):
        self.meno = meno
        self.vek = vek

osoba = Osoba("Miroslav", 30)

# Zobrazenie atrib√∫tov
print(osoba.meno)  # V√Ωstup: Miroslav

# Vymazanie atrib√∫tu
delattr(osoba, "meno")

# Pokus o pr√≠stup k vymazan√©mu atrib√∫tu vyvol√° chybu
# print(osoba.meno)  # Chyba: AttributeError
```

### Funkcia enumerate()
- Pou≈æ√≠va sa na z√≠skanie indexu a hodnoty v iterovateƒænom objekte.
- M√¥≈æe by≈• u≈æitoƒçn√° na z√≠skanie indexu v kombin√°cii s iterovan√≠m cez r√¥zne objekty, napr√≠klad pri spracovan√≠ zoznamu s r√¥znymi d√°tov√Ωmi typmi.
- Na z√≠skanie indexu a hodnoty zoznamu (alebo in√©ho iterovateƒæn√©ho objektu) poƒças iter√°cie. 
- T√°to funkcia vracia iter√°tor, ktor√Ω generuje dvojice, kde prv√Ω prvok je index a druh√Ω prvok je hodnota z iterovateƒæn√©ho objektu.
- Veƒæmi u≈æitoƒçn√°, keƒè potrebujete ma≈• pr√≠stup k indexu prvkov v iterovateƒænom objekte, najm√§ pri pr√°ci s cyklami, kde index a hodnota potrebuj√∫ by≈• spracovan√© s√∫ƒçasne.

**V√Ωhody:**
1. Funkcia enumerate() zjednodu≈°uje pr√°cu s indexmi poƒças iter√°cie, preto≈æe v√°m umo≈æ≈àuje z√≠ska≈• index a hodnotu s√∫ƒçasne bez potreby ruƒçne poƒç√≠ta≈• index.
2. Pou≈æitie enumerate() v iter√°ci√°ch je efekt√≠vne a k√≥d je ƒçistej≈°√≠, ne≈æ keby ste pou≈æili range() a len() na z√≠skanie indexu.

```python
# Z√≠skanie indexu a hodnoty zo zoznamu

animals = ['cat', 'dog', 'rabbit']
for index, animal in enumerate(animals):
    print(index, animal)

# Vytlaƒç√≠:
# 0 cat
# 1 dog
# 2 rabbit
```

```python
# Z√≠skanie indexu a hodnoty zo zoznamu s in√Ωm poƒçiatoƒçn√Ωm indexom
animals = ['cat', 'dog', 'rabbit']
for index, animal in enumerate(animals, start=1):
    print(index, animal)
# Vytlaƒç√≠:
# 1 cat
# 2 dog
# 3 rabbit
```

```python
# Pr√≠klad pre funkciu enumerate() s re≈•azcom
retezec = "Python"

for index, znak in enumerate(retezec):
    print(f"Index: {index}, Znak: {znak}")
```

### Funkcia exec() 
- Sp√∫≈°≈•a Python k√≥d z re≈•azca.
- Sl√∫≈æi na vykonanie (vyhodnotenie) re≈•azca, ktor√Ω obsahuje platn√Ω Python v√Ωraz. Tento re≈•azec sa spracuje ako k√≥d, ktor√Ω sa vykon√° v r√°mci aktu√°lneho kontextu (s√∫boru, skriptu, atƒè.). 
- Funkcia eval() je veƒæmi siln√°, ale m√¥≈æe by≈• nebezpeƒçn√°, ak sa pou≈æije s neoveren√Ωm vstupom, preto≈æe m√¥≈æe vykona≈• ak√Ωkoƒævek Python k√≥d.
- Pou≈æ√≠vanie eval() sa vo v≈°eobecnosti neodpor√∫ƒça, pokiaƒæ nie ste si ist√≠, ≈æe vstupy s√∫ bezpeƒçn√©. Ak potrebujete bezpeƒçne vyhodnoti≈• v√Ωrazy, zv√°≈æte pou≈æitie funkci√≠ ako ast.literal_eval(), ktor√° vyhodnocuje len liter√°ly, a teda je bezpeƒçnej≈°ia.

**V√Ωhody:**
1. eval() m√¥≈æe by≈• u≈æitoƒçn√Ω na dynamick√© vykon√°vanie k√≥du, ktor√Ω je definovan√Ω ako re≈•azec.
2. Umo≈æ≈àuje vyhodnocova≈• v√Ωrazy, ktor√© m√¥≈æu by≈• urƒçen√© a≈æ poƒças behu programu (napr. zo vstupu pou≈æ√≠vateƒæa, s√∫borov, atƒè.).

**Nev√Ωhody:**
1. Bezpeƒçnostn√© riziko: Ak sa pou≈æije s neoveren√Ωm vstupom (napr. od pou≈æ√≠vateƒæa), m√¥≈æe by≈• veƒæmi nebezpeƒçn√°, preto≈æe m√¥≈æe vykona≈• ƒæubovoƒæn√Ω k√≥d. Tento probl√©m sa d√° minimalizova≈• t√Ωm, ≈æe sa eval() pou≈æ√≠va iba s d√¥veryhodn√Ωmi d√°tami alebo sa obmedz√≠ pr√≠stup k nebezpeƒçn√Ωm funkci√°m (napr. importovanie kni≈æn√≠c).
2. N√°roƒçnos≈• na √∫dr≈æbu: Pou≈æ√≠vanie eval() m√¥≈æe sp√¥sobi≈•, ≈æe k√≥d bude ≈•a≈æ≈°ie ƒçitateƒæn√Ω a √∫dr≈æbov√Ω.

```python
# Pr√≠klad pre funkciu exec()
kod = """
def pozdrav():
    print("Ahoj, svet!")
"""

exec(kod)  # Definuje funkciu pozdrav()
pozdrav()  # V√Ωstup: Ahoj, svet!
```

### Funkcia filter() 
- Filtruje iterovateƒæn√Ω objekt na z√°klade funkcie, ktor√° vracia True alebo False.
- M√¥≈æe by≈• pou≈æit√° aj na filtrovanie re≈•azcov podƒæa urƒçit√Ωch podmienok.
- Sl√∫≈æi na filtrovanie prvkov z iterovateƒæn√©ho objektu podƒæa podmienky definovanej funkciou. T√°to funkcia vracia iter√°tor, ktor√Ω obsahuje len tie prvky, ktor√© splnia podmienky v zadanom filtri.
- Veƒæmi u≈æitoƒçn√°, keƒè potrebujete vybra≈• len urƒçit√© hodnoty zo sekvencie, ktor√© spƒ∫≈àaj√∫ preddefinovan√∫ podmienku.

**V√Ωhody:**
1. filter() je efekt√≠vny sp√¥sob na filtrovanie √∫dajov v iterovateƒæn√Ωch objektoch bez potreby manu√°lneho cyklu.
2. M√¥≈æete filtrova≈• d√°ta na z√°klade dynamick√Ωch podmienok, ktor√© m√¥≈æu by≈• definovan√© funkciami alebo lambda v√Ωrazy.
3. Funkcia vracia iter√°tor, ƒço znamen√°, ≈æe je pam√§≈•ovo efekt√≠vna a nevy≈æaduje vytv√°ranie celej k√≥pie filtrovan√Ωch d√°t.

**Nev√Ωhody:**
1. Funkcia filter() vracia iter√°tor, ƒço znamen√°, ≈æe je potrebn√© ju previes≈• na zoznam alebo in√∫ kolekciu, ak chcete pracova≈• s v√Ωsledkami priamo (napr. cez list(), tuple(), atƒè.).
2. M√¥≈æe by≈• menej ƒçitateƒæn√°, ak sa pou≈æ√≠va veƒæmi komplexn√° filtraƒçn√° funkcia.

```python
# Z√°kladn√© filtrovanie parnych cisel s vlastnou funkciou
zoznam = [1, 2, 3, 4, 5, 6]

# Funkcia, ktor√° vracia True pre p√°rne ƒç√≠sla
def je_parne(cislo):
    return cislo % 2 == 0

vysledok = filter(je_parne, zoznam)

# Prevod na zoznam a zobrazenie v√Ωsledku
print(list(vysledok))  # V√Ωstup: [2, 4, 6]
```

```python
numbers = [1, 2, 3, 4, 5, 6]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # Vytlaƒç√≠: [2, 4, 6]
```

```python
# Z√°kladn√© filtrovanie: Filtrovanie ƒç√≠sel zoznamu, ktor√© s√∫ v√§ƒç≈°ie ako 5:
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = filter(lambda x: x > 5, numbers)
print(list(result))  # Vytlaƒç√≠: [6, 7, 8]
```

```python
# Pr√≠klad pre funkciu filter() s re≈•azcami
zoznam_retezcov = ["python", "java", "c++", "ruby", "javascript"]

# Funkcia, ktor√° vracia True pre re≈•azce dlh≈°ie ako 4 znaky
def dlhy_retezec(retezec):
    return len(retezec) > 4

vysledok = filter(dlhy_retezec, zoznam_retezcov)

# Prevod na zoznam a zobrazenie v√Ωsledku
print(list(vysledok))  # V√Ωstup: ['python', 'javascript']
```

```python
# Filtrovanie re≈•azcov, ktor√© zaƒç√≠naj√∫ p√≠smenom "a":

words = ['apple', 'banana', 'avocado', 'cherry']
result = filter(lambda word: word.startswith('a'), words)
print(list(result))  # Vytlaƒç√≠: ['apple', 'avocado']
```

```python
# Funkcia vracaj√∫ca None (vyber√° iba pravdiv√© hodnoty): Filtrovanie iba t√Ωch hodn√¥t, ktor√© s√∫ True (nepr√°zdne, ne-nulov√©):
values = [0, 1, False, True, None, 'hello']
result = filter(None, values)
print(list(result))  # Vytlaƒç√≠: [1, True, 'hello']
```

### Funkcia frozenset() 
- Vytv√°ra nemenn√∫ (immutable) mno≈æinu.
- U≈æitoƒçn√° pri vytv√°ran√≠ nemenn√Ωch mno≈æ√≠n, ktor√© sa pou≈æ√≠vaj√∫ napr√≠klad na oper√°cie s nepozmie≈àateƒæn√Ωmi √∫dajmi.
- Na vytvorenie nezmenn√©ho (immutable) mno≈æiny. 
- Mno≈æiny (sets) v Pythone s√∫ kolekcie, ktor√© neobsahuj√∫ duplicitn√© prvky a umo≈æ≈àuj√∫ r√Ωchly pr√≠stup k prvkom. 
- Rozdiel medzi mno≈æinou (set()) a nepr√≠stupnou mno≈æinou (frozenset()) je ten, ≈æe frozenset je nemenn√Ω, tzn. po jeho vytvoren√≠ sa jeho obsah ned√° meni≈• (neprid√°vaj√∫ sa nov√© prvky, ani sa neodstra≈àuj√∫ existuj√∫ce).
- Ide√°lna, keƒè potrebujete pou≈æitie mno≈æiny s garantovanou nemennos≈•ou a bez rizika modifik√°ci√≠ poƒças pr√°ce s d√°tami.

**V√Ωhody:**
1. frozenset() je u≈æitoƒçn√©, keƒè potrebujete mno≈æinu, ktor√° by sa nemala meni≈•, ako napr√≠klad pri pou≈æit√≠ ako kƒæ√∫ƒç v slovn√≠ku alebo pri oper√°ci√°ch s mno≈æinami, kde nechcete povoli≈• ne√∫myseln√© zmeny.
2. Funguje podobne ako mno≈æina, ale poskytuje ochranu pred ne√∫myseln√Ωmi modifik√°ciami.

**Nev√Ωhody:**
1. Na rozdiel od be≈ænej mno≈æiny, nem√¥≈æete prida≈• ani odstr√°ni≈• prvky z frozenset po jeho vytvoren√≠.
2. Ak potrebujete meni≈• obsah, budete musie≈• pou≈æi≈• obyƒçajn√∫ mno≈æinu (set()).

```python
# Vytvorenie frozenset zo zoznamu:
mnozina = frozenset([1, 2, 3, 4, 5])

# Zobrazenie mno≈æiny
print(mnozina)  # V√Ωstup: frozenset({1, 2, 3, 4, 5})

# Pokus o prida≈• prvok sp√¥sob√≠ chybu, preto≈æe frozenset je nemenn√Ω
# mnozina.add(6)  # Chyba: AttributeError
```

```python
# Porovnanie s mno≈æinou (set()):
my_set = {1, 2, 3}
frozen_set = frozenset([1, 2, 3])

print(my_set == frozen_set)  # Vytlaƒç√≠: True, preto≈æe obsahuj√∫ rovnak√© prvky
```

```python
# Operacie s frozenset()
mnozina = frozenset([1, 2, 3, 4, 5])

# Pou≈æitie frozenset v kombin√°cii s oper√°ciami
dalsia_mnozina = frozenset([4, 5, 6, 7])

# Prienik dvoch mno≈æ√≠n
prienik = mnozina & dalsia_mnozina
print(prienik)  # V√Ωstup: frozenset({4, 5})

# Spojenie dvoch frozenset
zjednotenie = mnozina | dalsia_mnozina
print(zjednotenie)  # V√Ωstup: frozenset({1, 2, 3, 4, 5, 6, 7})
```

### Funkcia globals() 
- Vracia glob√°lny symbolov√Ω slovn√≠k (dictionary), ktor√Ω obsahuje v≈°etky glob√°lne premenn√© a funkcie v aktu√°lnom module.
- Tento slovn√≠k obsahuje n√°zvy premenn√Ωch a ich hodnoty, ktor√© s√∫ definovan√© v glob√°lnom rozsahu programu (mimo v≈°etk√Ωch funkci√≠ alebo tried). 
- Funkcia globals() umo≈æ≈àuje dynamicky pristupova≈• k t√Ωmto premenn√Ωm alebo ich upravova≈• poƒças behu programu.
- Veƒæmi siln√Ω n√°stroj, ktor√Ω umo≈æ≈àuje manipul√°ciu s glob√°lnymi d√°tami v Pythone, ale malo by sa s ≈àou zaobch√°dza≈• opatrne, aby sa predi≈°lo ne√∫myseln√Ωm zmen√°m v glob√°lnych premenn√Ωch.

**V√Ωhody:**
1. globals() je u≈æitoƒçn√©, ak potrebujete pristupova≈• alebo manipulova≈• s glob√°lnymi premenn√Ωmi poƒças behu programu.
2. Pom√°ha pri dynamick√Ωch oper√°ci√°ch, ako je evalu√°cia k√≥du alebo vytv√°ranie nov√Ωch premenn√Ωch alebo funkci√≠.

**Nev√Ωhody:**
1. Bezpeƒçnostn√© rizik√°: Pou≈æitie globals() na dynamick√∫ manipul√°ciu s glob√°lnymi premenn√Ωmi m√¥≈æe vies≈• k ne√∫myseln√Ωm chyb√°m alebo bezpeƒçnostn√Ωm probl√©mom, najm√§ ak pracujete s neoveren√Ωmi d√°tami.
2. K√≥d m√¥≈æe by≈• menej ƒçitateƒæn√Ω, ak sa premenn√© alebo funkcie manipuluj√∫ priamo cez globals().

```python
# Zobrazenie v≈°etk√Ωch glob√°lnych premenn√Ωch:

a = 10
b = 20

# Zobrazenie glob√°lnych premenn√Ωch
print(globals())  # V√Ωstup: {'__name__': '__main__', '__doc__': None, '__package__': None, 'a': 10, 'b': 20, ...}
```

```python
# Z√≠skanie hodnoty glob√°lnej premennej a pomocou globals():

a = 5
b = 10
print(globals()['a'])  # Vytlaƒç√≠: 5
```

```python
# Zmena hodnoty glob√°lnej premennej cez globals():

a = 5
globals()['a'] = 100
print(a)  # Vytlaƒç√≠: 100, preto≈æe sme upravili glob√°lnu premenn√∫ 'a' cez globals()
```

```python
def hello():
    return "Hello, World!"

print(globals()['hello']())  # Vytlaƒç√≠: Hello, World!
```

```python
# Pou≈æitie v dynamickom sp√∫≈°≈•an√≠ k√≥du

x = 1
y = 2

code = 'x + y'
result = eval(code, globals())
print(result)  # Vytlaƒç√≠: 3
```

### Funkcia iter() 
- Vytv√°ra iter√°tor z iterovateƒæn√©ho objektu.
- Sl√∫≈æi na z√≠skanie iter√°tora z iterovateƒæn√©ho objektu, ako s√∫ zoznamy, n-tice, re≈•azce alebo in√© objekty, ktor√© podporuj√∫ iterovanie. 
- Funkcia iter() vracia objekt, ktor√Ω implementuje met√≥dy __next__() a __iter__(), ƒço umo≈æ≈àuje iter√°ciu cez prvky objektu.
- Je veƒæmi u≈æitoƒçn√°, keƒè potrebujete manu√°lne spravova≈• iter√°ciu cez prvky objektu alebo ak chcete pr√≠stup k iter√°torom pre ƒèal≈°ie spracovanie

**V√Ωhody:**
1. iter() umo≈æ≈àuje explicitne z√≠ska≈• iter√°tor z iterovateƒæn√©ho objektu, ƒço je u≈æitoƒçn√©, ak chcete manu√°lne kontrolova≈• priebeh iter√°cie pomocou next().
2. M√¥≈æete pou≈æ√≠va≈• iter() aj na objekty, ktor√© nie s√∫ natrvalo v pam√§ti (napr. gener√°tory), ƒço umo≈æ≈àuje efekt√≠vnu spr√°vu pam√§te.

**Nev√Ωhody:**
1. Pri pou≈æit√≠ next() s iter√°torom je potrebn√© zaobch√°dza≈• opatrne, preto≈æe po vyƒçerpan√≠ iter√°tora (keƒè u≈æ nie s√∫ ƒèal≈°ie prvky na iterovanie) sa vyvol√° v√Ωnimka StopIteration, ƒço m√¥≈æe sp√¥sobi≈• ne≈æiaduce chyby, ak sa neprimerane nezachyt√≠.

```python
# Vytvorenie iter√°tora z zoznamu:

zoznam = [1, 2, 3, 4, 5]

# Vytvorenie iter√°tora
it = iter(zoznam)

# Iterovanie cez iter√°tor
print(next(it))  # V√Ωstup: 1
print(next(it))  # V√Ωstup: 2
```

```python
# Pou≈æitie v cykle for (automatick√° iter√°cia):
numbers = [1, 2, 3, 4, 5]
it = iter(numbers)
for num in it:
    print(num)
    
# Vytlaƒç√≠:
# 1
# 2
# 3
# 4
# 5
```

```python
# Ruƒçn√© pou≈æitie funkcie next() s iter√°torom:
words = ['apple', 'banana', 'cherry']
it = iter(words)
print(next(it))  # Vytlaƒç√≠: apple
print(next(it))  # Vytlaƒç√≠: banana
print(next(it))  # Vytlaƒç√≠: cherry
```

```python
# Pou≈æitie s pr√°zdnym objektom (vyvol√° StopIteration):
empty_list = []
it = iter(empty_list)

try:
    print(next(it))
except StopIteration:
    print("Koniec iter√°cie")  # Vytlaƒç√≠: Koniec iter√°cie
```

```python
# Pou≈æitie iter√°tora s funkciou next() a v√Ωchodzou hodnotou:

numbers = [1, 2, 3]
it = iter(numbers)
print(next(it, 'Koniec'))  # Vytlaƒç√≠: 1
print(next(it, 'Koniec'))  # Vytlaƒç√≠: 2
print(next(it, 'Koniec'))  # Vytlaƒç√≠: 3
print(next(it, 'Koniec'))  # Vytlaƒç√≠: Koniec (preto≈æe iter√°cia skonƒçila)
```

### Funkcia locals() 
- Vracia lok√°lny symbolov√Ω slovn√≠k, ktor√Ω obsahuje lok√°lne premenn√© v aktu√°lnej funkcii alebo bloku k√≥du.
- Slovn√≠k obsahuj√∫ci v≈°etky lok√°lne premenn√© a ich hodnoty v aktu√°lnom rozsahu (kontexte). 
- Tento rozsah m√¥≈æe by≈• funkcia, met√≥da, alebo in√Ω blok k√≥du, kde s√∫ premenn√© definovan√©. 
- Umo≈æ≈àuje dynamicky pristupova≈• k t√Ωmto premenn√Ωm alebo ich upravova≈• poƒças behu programu.
- Siln√Ω n√°stroj na manipul√°ciu a pristupovanie k lok√°lnym premenn√Ωm, ale mala by sa pou≈æ√≠va≈• opatrne, najm√§ pri dynamick√Ωch zmen√°ch hodn√¥t.

**V√Ωhody:**
1. locals() je u≈æitoƒçn√© na dynamick√© zistenie alebo manipul√°ciu s premenn√Ωmi v aktu√°lnom rozsahu.
2. M√¥≈æe by≈• pou≈æit√© v debugovan√≠ alebo pri generovan√≠ k√≥du, kde potrebujete pr√≠stup k aktu√°lnym hodnot√°m premenn√Ωch.

**Nev√Ωhody:**
1. Zmena hodn√¥t cez locals() nemus√≠ v≈ædy fungova≈•, najm√§ mimo funkcie, preto≈æe niektor√© lokalizovan√© premenn√© m√¥≈æu by≈• optimalizovan√© a neumo≈æ≈àuj√∫ priame zmeny.
2. Manipul√°cia s lok√°lnymi premenn√Ωmi cez locals() m√¥≈æe by≈• riskantn√°, preto≈æe m√¥≈æe sp√¥sobi≈• ne√∫myseln√© chyby v k√≥de.

```python
# Zobrazenie v≈°etk√Ωch lok√°lnych premenn√Ωch vo funkcii:

def funkcia():
    x = 5
    y = 10
    print(locals())  # V√Ωstup: {'x': 5, 'y': 10}

funkcia()

# print(locals())
```

```python
# Pou≈æitie locals() na pr√≠stup k lok√°lnym premenn√Ωm:
def my_function():
    a = 5
    b = 15
    print(locals()['a'])  # Vytlaƒç√≠: 5
    print(locals()['b'])  # Vytlaƒç√≠: 15

my_function()
```

```python
# Dynamick√© pridanie lok√°lnej premennej pomocou locals():

def my_function():
    a = 10
    locals()['new_var'] = 50
    print(locals()['new_var'])  # Vytlaƒç√≠: 50

my_function()
```

### Funkcia map() 
- Aplikuje funkciu na ka≈æd√Ω prvok iterovateƒæn√©ho objektu (napr. zoznamu).
- Pou≈æ√≠va na aplikovanie danej funkcie na ka≈æd√Ω prvok v iterovateƒænom objekte (ako je zoznam, n-tica, atƒè.). 
- Vracia iter√°tor, ktor√Ω obsahuje v√Ωsledky aplikovania funkcie na ka≈æd√Ω prvok vstupn√©ho iterovateƒæn√©ho objektu. 
- Na spracovanie d√°t, kde je potrebn√© pou≈æi≈• rovnak√∫ oper√°ciu na ka≈æd√Ω prvok v sekvencii.
- Veƒæmi u≈æitoƒçn√°, keƒè potrebujete aplikova≈• rovnak√∫ oper√°ciu na ka≈æd√Ω prvok v iterovateƒænom objekte, najm√§ pri pr√°ci s veƒæk√Ωmi mno≈æstvami d√°t, kde je efekt√≠vnos≈• pam√§te d√¥le≈æit√°.

**V√Ωhody:**
1. map() je efekt√≠vna na aplikovanie rovnak√Ωch oper√°ci√≠ na ka≈æd√Ω prvok v iterovateƒænom objekte bez potreby pou≈æitia explicitn√©ho cyklu for.
2. Vr√°ten√Ω iter√°tor je pam√§≈•ovo efekt√≠vny, preto≈æe spracov√°va d√°ta "na vy≈æiadanie", ƒço ≈°etr√≠ pam√§≈• pri pr√°ci s veƒæk√Ωmi d√°tov√Ωmi mno≈æinami.

**Nev√Ωhody:**
1. Funkcia map() vracia iter√°tor, tak≈æe je potrebn√© ju previes≈• na zoznam alebo in√∫ kolekciu (napr. cez list()), ak chcete pracova≈• s v√Ωsledkami priamo.
2. Ak je funkcia alebo pou≈æitie zlo≈æitej≈°ie, m√¥≈æe by≈• menej ƒçitateƒæn√© ako klasick√Ω cyklus for.

```python
# Z√°kladn√© pou≈æitie s jedn√Ωm iterovateƒæn√Ωm objektom: Aplikovanie funkcie na ka≈æd√Ω prvok zoznamu
zoznam = [1, 2, 3, 4, 5]

# Funkcia na zdvojn√°sobenie hodnoty
def zdvojnasobit(cislo):
    return cislo * 2

# Pou≈æitie map() na aplikovanie funkcie na ka≈æd√Ω prvok
vysledok = map(zdvojnasobit, zoznam)

# Prevod na zoznam a zobrazenie v√Ωsledku
print(list(vysledok))  # V√Ωstup: [2, 4, 6, 8, 10]
```

```python
numbers = [1, 2, 3, 4, 5]
result = map(lambda x: x * 2, numbers)
print(list(result))  # Vytlaƒç√≠: [2, 4, 6, 8, 10]
```

```python
# Aplikovanie funkcie na re≈•azce: Pre ka≈æd√© p√≠smeno re≈•azca aplikujeme funkciu, ktor√° ho prevedie na veƒæk√© p√≠smeno:
words = ['apple', 'banana', 'cherry']
result = map(str.upper, words)
print(list(result))  # Vytlaƒç√≠: ['APPLE', 'BANANA', 'CHERRY']
```

```python
# Pou≈æitie s definovanou funkciou: Pou≈æitie vlastnej funkcie na ka≈æd√∫ hodnotu zoznamu:

def square(x):
    return x ** 2

numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))  # Vytlaƒç√≠: [1, 4, 9, 16]
```

## üìä Modul csv ‚Äì ƒå√≠tanie CSV s√∫boru do zoznamu slovn√≠kov

**Dialekt (dialect):** s√∫bor nastaven√≠, ktor√© urƒçuj√∫, ako sa bud√∫ spracov√°va≈• CSV s√∫bory. Obsahuje ≈°pecifik√°cie pre r√¥zne parametre, ako s√∫:
- delimiter ‚Äì znak, ktor√Ω sa pou≈æ√≠va na oddelenie hodn√¥t (napr. ƒçiarka ,, bodkoƒçiarka ;).
- quotechar ‚Äì znak, ktor√Ω sa pou≈æ√≠va na ohraniƒçenie hodn√¥t obsahuj√∫cich oddeƒæovaƒçe alebo ≈°peci√°lne znaky (napr. √∫vodzovky ").
- escapechar ‚Äì znak, ktor√Ω sa pou≈æ√≠va na √∫nik pred ≈°peci√°lnymi znakmi (napr. sp√§tn√© lom√≠tko \).
- lineterminator ‚Äì znak (alebo sekvencia), ktor√Ω sa pou≈æ√≠va na ukonƒçenie riadku.
- skipinitialspace ‚Äì ak je nastaven√Ω na True, ignoruj√∫ sa medzery po oddeƒæovaƒçoch.

1. csv.reader(csvfile, delimiter=',') ‚Äì ƒç√≠ta CSV s√∫bor a vracia iterator na riadky (predpoklad√° sa, ≈æe s√∫ hodnoty oddelen√© ƒçiarkami)
2. csv.writer(csvfile, delimiter=',') ‚Äì vytv√°ra objekt na z√°pis do CSV s√∫boru
3. csv.DictReader(csvfile) ‚Äì ƒç√≠ta CSV s√∫bor a vracia riadky ako slovn√≠ky, kde kƒæ√∫ƒçe s√∫ n√°zvy stƒ∫pcov
4. csv.DictWriter(csvfile, fieldnames) ‚Äì zapisuje riadky do CSV s√∫boru ako slovn√≠ky, kde kƒæ√∫ƒçe s√∫ n√°zvy stƒ∫pcov
5. csv.field_size_limit(limit) ‚Äì nastavuje limit veƒækosti poƒæa pre ƒç√≠tanie CSV s√∫boru
6. csv.register_dialect(name, delimiter, quotechar, escapechar) ‚Äì zaregistruje nov√Ω dialekt pre CSV form√°t
7. csv.unregister_dialect(name) ‚Äì odregistrov√° dialekt podƒæa n√°zvu
8. csv.get_dialect(name) ‚Äì z√≠ska parametre pre zaregistrovan√Ω dialekt

üìå √öloha: Form√°tova≈• d√°tum

1. %Y ‚Äì Rok (4 ƒç√≠slice, napr. 2025)
2. %y ‚Äì Rok (2 ƒç√≠slice, napr. 25)
3. %m ‚Äì Mesiac (2 ƒç√≠slice, napr. 03)
4. %b ‚Äì Skr√°ten√Ω n√°zov mesiaca (Jan, Feb, Mar, ...)
5. %B ‚Äì N√°zov mesiaca (Janu√°r, Febru√°r, ...)
6. %d ‚Äì De≈à mesiaca (2 ƒç√≠slice, napr. 09)
7. %j ‚Äì De≈à v roku (3 ƒç√≠slice, napr. 009)
8. %A ‚Äì N√°zov d≈àa v t√Ω≈ædni (Pondelok, Utorok, ...)
9. %a ‚Äì Skr√°ten√Ω n√°zov d≈àa v t√Ω≈ædni (Pon, Uto, ...)
10. %w ‚Äì De≈à v t√Ω≈ædni (0-6, kde 0 je nedeƒæa)
11. %H ‚Äì Hodina v 24-hodinovom form√°te (00-23)
12. %I ‚Äì Hodina v 12-hodinovom form√°te (01-12)
13. %M ‚Äì Min√∫ty (00-59)
14. %S ‚Äì Sekundy (00-59)
15. %p ‚Äì AM alebo PM (len pre 12-hodinov√Ω form√°t)
16. %f ‚Äì Mikrosekundy (6 ƒç√≠slic)
17. %z ‚Äì Posun od UTC v ƒçasovom p√°sme (napr. +0200)
18. %Z ‚Äì N√°zov ƒçasov√©ho p√°sma (napr. CET, UTC)
19. %X ‚Äì ƒåas vo form√°te (HH:MM:SS)
20. %D ‚Äì Kr√°tky d√°tum (MM/DD/YY)
21. %F ‚Äì D√°tum vo form√°te YYYY-MM-DD
22. %T ‚Äì ƒåas vo form√°te HH:MM:SS

```python
from datetime import datetime

datum = datetime.now()
# 19.03.2025 10:08:22
print(datum.strftime("%d.%m.%Y %H:%M:%S"))

# Rok-Mesiac-De≈à (2025-03-18)
print(datum.strftime("%Y-%m-%d"))
        
# De≈à/Mesiac/Rok (18/03/2025)
print(datum.strftime("%d/%m/%Y"))
        
# N√°zov mesiaca De≈à, Rok (Marec 18, 2025)
print(datum.strftime("%B %d, %Y")) 
        
# Skr√°ten√Ω n√°zov d≈àa, De≈à Skr√°ten√Ω n√°zov mesiaca Rok (Uto, 18 Mar 2025)        
print(datum.strftime("%a, %d %b %Y"))  
        
# Hodina:min√∫ty AM/PM (03:45 PM)
print(datum.strftime("%I:%M %p"))

# Rok-Mesiac-De≈à Hodina:Min√∫ty:Sekundy (2025-03-18 15:45:30)
print(datum.strftime("%Y-%m-%d %H:%M:%S"))

# N√°zov d≈àa, N√°zov mesiaca De≈à, Rok (Pondelok, Marec 18, 2025)
print(datum.strftime("%A, %B %d, %Y"))

# Hodina:Min√∫ty:Sekundy (15:45:30)
print(datum.strftime("%H:%M:%S"))

# D√°tum vo form√°te YYYY-MM-DD (2025-03-18)
print(datum.strftime("%F"))

#  ƒåas vo form√°te HH:MM:SS (15:45:30)
print(datum.strftime("%T"))
```

5. defaultdict() ‚Äì slovn√≠k, ktor√Ω automaticky inicializuje hodnotu pre neexistuj√∫ce kƒæ√∫ƒçe

üìå √öloha: Pou≈æitie namedtuple

```python
from collections import namedtuple

# Vytvorenie namedtuple pre bod
Point = namedtuple('Point', ['x', 'y'])

# Vytvorenie in≈°tancie namedtuple
p = Point(10, 20)

# Zobrazenie hodn√¥t
print(p)
# V√Ωstup: Point(x=10, y=20)

# Pr√≠stup k hodnot√°m pomocou pomenovan√Ωch pol√≠
print(p.x)  # V√Ωstup: 10
print(p.y)  # V√Ωstup: 20
```

üìå √öloha: Pou≈æitie ChainMap

```python
from collections import ChainMap

# Vytvorenie dvoch slovn√≠kov
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}

# Kombinovanie slovn√≠kov do ChainMap
chain_map = ChainMap(dict1, dict2)

# Zobrazenie ChainMap
print(chain_map)
# V√Ωstup: ChainMap({'a': 1, 'b': 2}, {'b': 3, 'c': 4})

# Pr√≠stup k hodnot√°m (pozor, ChainMap prioritizuje prv√Ω slovn√≠k)
print(chain_map['b'])  # V√Ωstup: 2 (z dict1, preto≈æe je na prvom mieste)
print(chain_map['c'])  # V√Ωstup: 4 (z dict2)
```

üìå √öloha: Pou≈æitie Counter na spoƒç√≠tanie prvkov v zozname

```python
from collections import Counter
zoznam = ["a", "b", "a", "c", "b", "a"]
print(Counter(zoznam))  # {'a': 3, 'b': 2, 'c': 1}
```

üìå √öloha: Pou≈æitie OrderedDict

```python
from collections import OrderedDict

# Vytvorenie OrderedDict s pridan√≠m polo≈æiek
ordered_dict = OrderedDict()
ordered_dict['a'] = 1
ordered_dict['b'] = 2
ordered_dict['c'] = 3
ordered_dict['d'] = 4
ordered_dict['e'] = 5

# Zobrazenie OrderedDict
print(ordered_dict)
# V√Ωstup: OrderedDict([('a', 1), ('b', 2), ('c', 3), ('d', 4), ('e', 5)])

# Zmena poradia kƒæ√∫ƒçov
ordered_dict.move_to_end('b')
print(ordered_dict)
# V√Ωstup: OrderedDict([('a', 1), ('c', 3), ('d', 4), ('e', 5), ('b', 2)])
```

üìå √öloha: Pou≈æitie UserDict

```python
from collections import UserDict

# Vytvorenie vlastn√©ho slovn√≠ka pomocou UserDict
class CustomDict(UserDict):
    def __setitem__(self, key, value):
        print(f"Prid√°vam {key}: {value}")
        super().__setitem__(key, value)

# Pou≈æitie CustomDict
custom_dict = CustomDict()
custom_dict['x'] = 10
custom_dict['y'] = 20
print(custom_dict)
# V√Ωstup:
# Prid√°vam x: 10
# Prid√°vam y: 20
# {'x': 10, 'y': 20}
```

üìå √öloha: Pou≈æitie defaultdict na automatick√© vytv√°ranie hodn√¥t

```python
from collections import defaultdict
dd = defaultdict(int)
dd["a"] += 1
print(dd)  # {'a': 1}
```

üìå √öloha: Pou≈æitie deque na efekt√≠vne prid√°vanie a odoberanie prvkov

```python
from collections import deque
q = deque([1, 2, 3])
q.appendleft(0)
q.append(4)
print(q)  # deque([0, 1, 2, 3, 4])
```

```python
# zoznam adres√°rov, ktor√© Python pou≈æ√≠va na hƒæadanie modulov.

import sys

# Zobrazi≈• v≈°etky adres√°re, kde Python hƒæad√° moduly
print("Zoznam adres√°rov v sys.path:")
for path in sys.path:
    print(path)
```

```python
# Ak√° platforma je aktu√°lne pou≈æ√≠van√°.
import sys

# Zobrazi≈• platformu, na ktorej Python be≈æ√≠
print(f"Platforma, na ktorej be≈æ√≠ Python: {sys.platform}")
```

```python
# Ako zisti≈• veƒækos≈• objektu v pam√§ti v bajtoch.

import sys

# Vytvorenie r√¥znych objektov
text = "Hello, World!"
cislo = 123
zoznam = [1, 2, 3, 4, 5]

# Zobrazi≈• veƒækosti objektov v bajtoch
print(f"Veƒækos≈• textu: {sys.getsizeof(text)} bajtov")
print(f"Veƒækos≈• ƒç√≠sla: {sys.getsizeof(cislo)} bajtov")
print(f"Veƒækos≈• zoznamu: {sys.getsizeof(zoznam)} bajtov")
```
